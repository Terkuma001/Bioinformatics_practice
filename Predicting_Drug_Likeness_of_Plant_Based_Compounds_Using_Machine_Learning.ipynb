```python
# ✅ Now safe to import
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create or Load Your Dataset
# Example dataset with plant-derived compounds and SMILES strings
data = {
    'Compound': ['Quercetin', 'Caffeine', 'Eugenol', 'Curcumin', 'Linalool'],
    'SMILES': [
        'C1=CC(=C(C=C1O)O)C2=C(C(=O)C3=C(C=C(C=C3O2)O)O)O',  # Quercetin
        'CN1C=NC2=C1C(=O)N(C(=O)N2C)C',                      # Caffeine
        'CC1=CC(=CC=C1O)C=CC=C',                             # Eugenol
        'CC(=O)CC(=O)/C=C/C1=CC(=C(C=C1)O)O',                # Curcumin
        'CC(C)CCC=C(C)C'                                     # Linalool
    ]
}

df = pd.DataFrame(data)
df
```

```python
# Compute Lipinski Descriptors
# Function to calculate Lipinski descriptors
def compute_lipinski(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        mw = Descriptors.MolWt(mol)
        logp = Descriptors.MolLogP(mol)
        hbd = Descriptors.NumHDonors(mol)
        hba = Descriptors.NumHAcceptors(mol)
        return pd.Series([mw, logp, hbd, hba])
    else:
        return pd.Series([None]*4)

# Add descriptors to DataFrame
df[['MW', 'LogP', 'HDonors', 'HAcceptors']] = df['SMILES'].apply(compute_lipinski)
df
```

```python
# Label Compounds as Drug-like or Not
# Using Lipinski's Rule of Five
def label_druglike(row):
    return int((row['MW'] < 500) and (row['LogP'] < 5) and (row['HDonors'] <= 5) and (row['HAcceptors'] <= 10))

df['DrugLike'] = df.apply(label_druglike, axis=1)
df
```

```python
# Train a Machine Learning Model
# Feature matrix and label
X = df[['MW', 'LogP', 'HDonors', 'HAcceptors']]
y = df['DrugLike']

# Split into train-test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluation
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Visualize Results
# Pairplot
sns.pairplot(df, hue='DrugLike', vars=['MW', 'LogP', 'HDonors', 'HAcceptors'])
plt.suptitle('Drug-Likeness Feature Distribution', y=1.02)
plt.show()
```

```python
# Create an Interactive SMILES Input Form in Colab
#@title 🔍 Drug-Likeness Prediction Form
from IPython.display import display
import ipywidgets as widgets

# Create input box for SMILES
smiles_input = widgets.Text(
    value='',
    placeholder='Enter SMILES here',
    description='SMILES:',
    disabled=False
)

# Display input box
display(smiles_input)

# Button to trigger prediction
predict_button = widgets.Button(description="Predict Drug-Likeness")
display(predict_button)

# Output area
output = widgets.Output()
display(output)

# Define prediction function
def on_button_clicked(b):
    output.clear_output()
    with output:
        smiles = smiles_input.value
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            features = [
                Descriptors.MolWt(mol),
                Descriptors.MolLogP(mol),
                Descriptors.NumHDonors(mol),
                Descriptors.NumHAcceptors(mol)
            ]
            prediction = model.predict([features])[0]
            print(f"Prediction: {'✅ Drug-like' if prediction == 1 else '❌ Not Drug-like'}")
            print(f"Descriptors → MW: {features[0]:.2f}, LogP: {features[1]:.2f}, HDonors: {features[2]}, HAcceptors: {features[3]}")
        else:
            print("❗ Invalid SMILES string. Please check and try again.")

# Connect button to function
predict_button.on_click(on_button_clicked)
```

```python
import gradio as gr
from rdkit import Chem
from rdkit.Chem import Descriptors
import joblib

# Load your model
model = joblib.load("drug_likeness_model.pkl")

def predict_drug_likeness(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        features = [
            Descriptors.MolWt(mol),
            Descriptors.MolLogP(mol),
            Descriptors.NumHDonors(mol),
            Descriptors.NumHAcceptors(mol)
        ]
        pred = model.predict([features])[0]
        return "✅ Drug-like" if pred == 1 else "❌ Not Drug-like"
    else:
        return "❗ Invalid SMILES"

interface = gr.Interface(fn=predict_drug_likeness,
                         inputs="text",
                         outputs="text",
                         title="Terkuma Saaondo's Drug-Likeness Predictor",

                         description="Enter a SMILES string to predict if it's drug-like.")

interface.launch()
```
